```python
import shapefile as shp
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
import matplotlib.font_manager as font_manager
import numpy as np
import os.path as path
from matplotlib.patches import Rectangle
plt.style.use('ggplot')
#read doc here: https://pypi.python.org/pypi/pyshp
def make_poly(data):
    return Polygon(data, True)
    
def make_shape(sh, **kwarg):
    patches = []
    for x,y in zip(sh.parts, sh.parts[1:]):
        poly = make_poly(sh.points[x:y])
        # pprint(sh.points[x:y])
        patches.append(poly)
    patches.append(make_poly(sh.points[sh.parts[-1]:]))
    return PatchCollection(patches, **kwarg)

def clean_record(sf):
    records = sf.records()
    for rec in records:
        rec[1] = rec[1].decode('TIS620')
    return records
```

```python
sf = shp.Reader("gis/Changwat_77P.shp")

shapes = sf.shapes()
records = clean_record(sf)
```

```python
def plot_me(prev, records, shapes, title, fname=None):
    plt.figure(figsize=(8.27,11.69))
    ncolor = 5
    cmap = plt.cm.get_cmap('autumn_r', ncolor)
    scale = 1.5 / (1-1/ncolor)
    # n*w = scale
    # 1.5 = scale - w -> 1.5 = scale - scale/n = scale(1-1/n) --> scale = 1.5/(1-1/n)
    for shape, rec, data in zip(shapes, records, prev):
        code = rec[0]
        props = dict(edgecolor='k', facecolor='w')
        props['facecolor']=cmap(data*100/scale)

        plt.gca().add_collection(make_shape(shape, alpha=0.4, **props))

    plt.axes().set_aspect('equal', 'datalim')
    plt.gca().autoscale_view(True,True,True)
    plt.axis('off')
    plt.gca().autoscale_view(True,True,True)
    left_cutoff = np.linspace(0,scale,ncolor+1)[:-1]

    patches = [
        Rectangle((0, 0), 1, 1, fc=cmap(x/scale)) for x in left_cutoff+0.001]
    labels = ['>%.1f%%'%x for x in left_cutoff]
    plt.legend(patches, labels, loc='lower right', prop={'size':18})
    plt.title(title, fontdict={'fontsize':18})
    #plt.savefig(fname)
    
```

```python
prev = [np.random.random_sample()*0.015 for rec in records]
```

```python
plot_me(prev, records, shapes, title='Title')
plt.show()
```
